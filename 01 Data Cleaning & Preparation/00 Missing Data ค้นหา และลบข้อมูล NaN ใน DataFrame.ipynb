## Missing Data

เมื่อมีข้อมูลจำนวนมาก มีโอกาสที่ข้อมูลของเราจะมีไม่ครบซึ่งเราจะต้องมีการจัดการข้อมูลเหล่านี้ก่อน หากมีผลต่อการใช้งาน

## นำเข้าเครื่องมือ

#### นำเข้าเครื่องมือ pandas

```python
import pandas as pd
```

## ดึงข้อมูลจาก excel

#### ดึงข้อมูลที่จะทำงานด้วย excel ตั้งค่า column,row กับข้อมูลจากไฟล์ condo.xlsx

```python
table = pd.read_excel(r'C:\Users\acer\Desktop\Ultimate Python Excel Automation\Data Cleaning & Preparation\condo.xlsx')
table = table.set_index('ชื่อคอนโด')
table.head(5)
```

## หาข้อมูล nan

#### การหาข้อมูลที่ขาดหายด้วยคำสั่ง .isnull()

```python
table.isna()
```

## ตรวจข้อมูลขาดหายด้วยการนับ

#### การนับจำนวนด้วย .sum() ค่า True จะมีค่าเป็น 1 False จะเป็น 0

```python
table.isnull().sum(axis=1)
```

## ลบ nan

#### ลบข้อมูลที่ขาดหาย ด้วยคำสั่ง .dropna() ทุกช่อง หรือบางช่องด้วย parameter how= any หรือ all

```python
table.dropna(how='all')
```

#### ลบข้อมูลที่มีอย่างน้อย... ตัว ด้วย paramater thresh=

```python
table.dropna(thresh=4)
```
